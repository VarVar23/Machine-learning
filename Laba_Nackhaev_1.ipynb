Преобразовать JSON-файл в CSV-файл

```python
import json
import csv
```

```python
headers = ['item', 'country', 'year', 'sales']

with open('sales.json', 'r') as file:
    data_json = json.load(file)

with open('sales.csv', 'w') as csv_f: 
    new_csv_f = csv.DictWriter(csv_f, fieldnames = headers) 
    new_csv_f.writeheader() 
    
    for item in data_json:
        countrys = item['sales_by_country']
        
        for country in countrys:
            years = countrys[country]
            
            for year in years: 
                new_csv_f.writerow({'item':item['item'],'country':country,'year':year,'sales':item['sales_by_country'][country][year]})
```

4. Написать функцию transformation_plot, принимающую на вход набор двумерных точек и квадратную матрицу размером 2x2. Отрисуйте на одном графике оригинальные точки, на втором - точки после преобразования при помощи матрицы. Отобразите точки таким образом, чтобы было понятно, какая из точек на первом графике соответствует какой точке на втором.

```python
import random
```

```python
def transformation_plot(dots, a):
    c = dots[:,0]
    new_dots = np.dot(dots, a.T)
    plt.scatter(dots[:,0], dots[:,1], c=c)
    plt.show()
    plt.scatter(new_dots[:,0], new_dots[:,1], c=c)
    plt.show()
```

```python
dots = np.random.randint(0, 10, size = 50)
dots.resize(25,2)
a = np.array([[1,2],[3,4]])
transformation_plot(dots, a)
```

6. Дано множество векторов V размерности d. Дан вектор q такой же размерности. Определить:
а) Пропорцию векторов v в V, для которых угол(v,q) < 90 градусов
б) Пропорцию векторов v в V, для которых угол(v,q) < 30 градусов


```python
d = random.randrange(2,6,1)
k = random.randrange(2,10,1)

V = np.random.uniform(0,100,size=(k,d))
q = np.random.uniform(0,100,size=d)

cos = np.dot(V,q)/((np.sqrt(np.sum(q**2)))*(np.sqrt(np.sum(V**2, axis=1))))
angle_array = np.arccos(cos)*180/np.pi

sum_less_ninty = np.sum(angle_array < 90)
sum_less_thirty = np.sum(angle_array < 30)

print ('пропорция векторов v в V, для которых угол(v,q) < 90 градусов = ', sum_less_ninty/k)
print ('пропорция векторов v в V, для которых угол(v,q) < 30 градусов = ', sum_less_thirty/k)
```
